```python
from scipy import stats
x = [6,8,9,10,12,15,18,20,25]
y = [100,150,160,190,200,250,340,450,466]
len(x),len(y)
```

```python
stats.linregress(x,y)
```

```python
slope,intercept,r,p,stderr = stats.linregress(x,y)

def fx(x):
    return slope*x+intercept # mx+c

model = []
for var in x:
    r = fx(var)
    model.append(r)

model
```

```python
plt.plot(x,model,c='r',marker='o')
```

```python
iris = sn.load_dataset('iris')
iris = iris[['petal_length','petal_width']]

X = iris['petal_length']
Y = iris['petal_width']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.4,random_state=23)
```

```python
x_train = np.array(x_train).reshape(-1,1)
y_train = np.array(y_train).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

model_obj = LinearRegression()
print(model_obj)
```

```python
model_obj.fit(x_train,y_train)
```

```python
m = model_obj.intercept_
c = model_obj.coef_
y_pred_train = m * x_train + c
y_pred_train = model_obj.predict(x_train)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,y_pred_train,color='r',marker='o')
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```

```python
from sklearn.metrics import r2_score
r2_score(y_train,y_pred_train)
```
